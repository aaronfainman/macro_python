```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import dateutil.parser
```

## READ IN SHARE PRICES

```python
df_voda = pd.read_csv('vodacom_share_price.csv', sep=';')
df_voda['year'] = 0
df_voda['month'] = 0
for i in range(len(df_voda)):
    df_voda['year'][i] = dateutil.parser.isoparse(df_voda['time'][i]).year
    df_voda['month'][i] = dateutil.parser.isoparse(df_voda['time'][i]).month


print("voda")
print(df_voda.head())
df_mtn = pd.read_csv('mtn_share_price.csv', sep=';')
print("\n mtn")
print(df_mtn.head())
df_telkom = pd.read_csv('telkom_share_price.csv', sep=';')
print("\n telkom")
print(df_telkom.head())
```

## READ IN SECTOR INDICES

```python
df_alsi = pd.read_csv('alsi.csv', sep=';')
print("ALSI")
print(df_alsi.head())

df_tel_index = pd.read_csv('j657_index_mobile_telecom.csv', sep=';')
print(" ")
print("J657 Mobile Telecoms Index")
print(df_tel_index.head())
```

## READ IN ECONOMIC INDICATORS

```python
df_gdp = pd.read_csv('world_gdp.csv', sep=';', header = None)
print("GDP: 2000 - 2019")
df_gdp = df_gdp.iloc[:, 41:61]
df_gdp = df_gdp.transpose()
df_gdp.columns = ['year','gdp']
print(df_gdp)

df_gdp_growth = pd.read_csv('gdp_growth_rate.csv', sep=';', header = None)
print("\n \n GDP Growth Rate: 2000 - 2021 (forecast)")
df_gdp_growth = df_gdp_growth.iloc[:, 41:62]
df_gdp_growth = df_gdp_growth.transpose()
df_gdp_growth.columns = ['year','gdp']
print(df_gdp_growth)
```

## PLOT DATA 

```python
df_gdp.plot(x = 'year', y = 'gdp', kind = 'line')
plt.title('GDP')
plt.show()
```

```python
df_gdp_growth.plot(x = 'year', y = 'gdp', kind = 'line')
plt.title('GDP')
plt.show()
```
